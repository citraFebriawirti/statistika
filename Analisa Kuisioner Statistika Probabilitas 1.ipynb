```python
import matplotlib.pyplot as plt # import package untuk data visualization
import pandas as pd # import package untuk data manipulation
import seaborn as sns

url = 'https://raw.githubusercontent.com/citraFebriawirti/statistika/main/stat_prop_satu%20(1).csv'

result = pd.read_csv(url) # membaca data berupa format comma separated value
```

```python
result.head() # melihat 5 data teratas
```

```python
drop = ['Timestamp','Email Address','No.Bp ']
result.drop(drop,axis=1,inplace=True)
result.drop(1,axis=0,inplace=True)
result = result.reset_index(drop=True)
result.set_index('Nama ',inplace=True)
result.head()
```

```python
nl = round(result.isnull().sum()/len(result)*100,2)
print(nl)
```

Dikarenakan null value pada kolom keempat lebih dari 50% dan tidak memiliki relasi terhadap kolom lainnya, maka data pada kolom ke-empat bisa dihapus

```python
result.drop(result.columns[3],axis=1,inplace=True)
nl = round(result.isnull().sum()/len(result)*100,2)
print(nl)
```

```python
def hasil(angka):
  plt.figure(figsize=(15,8))
  ax = sns.countplot(x=result.columns[angka],
                     data=result)
  plt.xticks(rotation=90)
  plt.xlabel("Jawaban")
  plt.title(result.columns[angka])
  plt.show()
```
